```python
import numpy as np
import matplotlib.pyplot as plt
import tclab
import time
# pip install gekko
from gekko import GEKKO
from numpy import random


%matplotlib widget
#%matplotlib notebook

t_sim = 300  # Number of second time points (5 min)

# simulation
m = GEKKO()
m.time = np.linspace(0,t_sim,t_sim+1) # n points and n+1 seconds results in 1 point per second

#TC = m.Var(23)
TC = m.Var(25.1) # Initial temperature

m.Equation(120*TC.dt()==(23-TC)+0.8*50)
m.options.IMODE = 4 # dynamic simulation
m.solve(disp=False)

# Plot results
fig = plt.figure(1)
plt_ax = plt.axes()

plt.plot(m.time,TC,'b-',label='Simulated')

plt_ax.set_ylim(20,60)
#plt_ax.autoscale(enable=True, axis = 'y')

##########################################
T1 = np.empty(t_sim)
T1[:] = np.nan
#print(T1)
T1_time = np.empty(t_sim)
T1_time[:] = np.nan
#print(T1_time)
#plt.plot(m.time,T1,'r.',label='Measured')
line1, = plt.plot(T1_time,T1,'r.',label='Measured')
##########################################
plt.ylabel('Temperature (degC)')
plt.xlabel('Time (sec)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Actual measured data
lab = tclab.TCLab()
#T1 = [lab.T1] # Old-style list based plotting...
lab.Q1(50)

#plt_ax = plt.gca()
T1[:] = np.nan       # Reset the measured temperatures to NaN's
T1_time[:] = np.nan

plt_ax.set_ylim(15,60)

for i in range(t_sim):
    time.sleep(1)
    print(lab.T1)
    T1[i] = lab.T1
    #Sim_Temp = random.rand() + 24
    #print(Sim_Temp)
    #T1[i] = Sim_Temp
    T1_time[i] = i
    #print(i)
    #print(T1)
    #print(T1_time)
    line1.set_ydata(np.array(T1))
    line1.set_xdata(np.array(T1_time))
    plt_ax.set_title("{0} of {1} Points Measured".format(i+1,t_sim))
    #plt_ax.autoscale(enable=True, axis = 'y')
    #plt_ax.autoscale()
    #plt.autoscale(axis = 'y')
    fig.canvas.draw()
    fig.canvas.flush_events()

lab.close()
```
